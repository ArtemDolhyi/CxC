```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (12, 8)
sns.set_theme(style="whitegrid")

df = pd.read_csv("SAP Datasets.csv")

id_columns = [
    'Country Name',
    'Country Code',
    'Indicator Name',
    'Topic',
    'short description',
    'long description',
    'Indicator Code',
    'Unit of measure'
]

year_cols = [str(y) for y in range(2000, 2024) if str(y) in df.columns]

df_long = pd.melt(
    df,
    id_vars=id_columns,
    value_vars=year_cols,
    var_name="Year",
    value_name="Value"
)

df_wide = df_long.pivot_table(
    index=["Country Name", "Year"],
    columns="Indicator Name",
    values="Value",
    aggfunc="mean"
).reset_index()

print("df_wide shape:", df_wide.shape)
df_wide.head(5)
```

```python
chosen_indicators = [
    "Access to clean fuels and technologies for cooking (% of population)",
    "Access to electricity (% of population)",
    "Adequacy of social protection and labor programs (% of total welfare of beneficiary households)",
    "Adjusted net national income per capita (current US$)",
    "Adjusted savings: education expenditure (current US$)",
    "Adolescent fertility rate (births per 1,000 women ages 15-19)",
    "Adolescents out of school (% of lower secondary school age)"
]

core_cols = ["Country Name", "Year"]
valid_chosen = [col for col in chosen_indicators if col in df_wide.columns]
df_mpi = df_wide[core_cols + valid_chosen].copy()

print("Chosen indicators found:", valid_chosen)
print("df_mpi shape:", df_mpi.shape)
df_mpi.head(5)
```

```python
for col in valid_chosen:
    df_mpi[col] = pd.to_numeric(df_mpi[col], errors="coerce")

missing_counts = df_mpi[valid_chosen].isnull().sum()
print("Missing values in chosen indicators:")
print(missing_counts)

df_mpi.dropna(axis=0, how="all", subset=valid_chosen, inplace=True)

print("df_mpi shape after dropping rows with all indicator columns = NaN:", df_mpi.shape)
df_mpi.head(5)
```

```python
from sklearn.preprocessing import MinMaxScaler

positive_inds = [
    "Access to clean fuels and technologies for cooking (% of population)",
    "Access to electricity (% of population)",
    "Adequacy of social protection and labor programs (% of total welfare of beneficiary households)",
    "Adjusted net national income per capita (current US$)",
    "Adjusted savings: education expenditure (current US$)"
]

negative_inds = [
    "Adolescent fertility rate (births per 1,000 women ages 15-19)",
    "Adolescents out of school (% of lower secondary school age)"
]

df_scaled = df_mpi.copy()

scaler = MinMaxScaler()
df_scaled[valid_chosen] = scaler.fit_transform(df_scaled[valid_chosen])

for col in negative_inds:
    df_scaled[col] = 1 - df_scaled[col]

print("Scaled & Inverted sample rows:")
df_scaled.head(10)
```

```python
df_index = df_scaled.copy()

df_index["MPI_Composite"] = df_index[valid_chosen].mean(axis=1)

df_index.head(10)
```

```python
year_to_analyze = "2020"

df_year = df_index[df_index["Year"] == year_to_analyze].copy()
df_year.dropna(subset=["MPI_Composite"], inplace=True)

df_year.sort_values("MPI_Composite", ascending=False, inplace=True)

top_10 = df_year.head(10)
plt.figure(figsize=(10,6))
sns.barplot(x="MPI_Composite", y="Country Name", data=top_10, palette="viridis")
plt.xlabel("MPI Composite (0-1)")
plt.ylabel("Country Name")
plt.title(f"Top 10 Countries by MPI_Composite in {year_to_analyze}")
plt.show()

bottom_10 = df_year.tail(10)
plt.figure(figsize=(10,6))
sns.barplot(x="MPI_Composite", y="Country Name", data=bottom_10, palette="rocket")
plt.xlabel("MPI Composite (0-1)")
plt.ylabel("Country Name")
plt.title(f"Bottom 10 Countries by MPI_Composite in {year_to_analyze}")
plt.show()
```

```python
corrs = {}
for col in valid_chosen:
    corrs[col] = df_index[col].corr(df_index["MPI_Composite"])

sorted_corrs = sorted(corrs.items(), key=lambda x: abs(x[1]), reverse=True)
print("Correlation of each indicator with MPI_Composite:")
for ind, val in sorted_corrs:
    print(f"{ind}: {val:.3f}")

from sklearn.decomposition import PCA

pca = PCA()
X = df_index[valid_chosen].dropna().values
pca.fit(X)

variance_ratios = pca.explained_variance_ratio_
print("\nPCA Explained Variance Ratios:")
for i, var_ratio in enumerate(variance_ratios):
    print(f"PC{i+1}: {var_ratio:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

plt.rcParams["figure.figsize"] = (12, 8)
sns.set_theme(style="whitegrid")


chosen_indicators = [
    "Access to clean fuels and technologies for cooking (% of population)",
    "Access to electricity (% of population)",
    "Adequacy of social protection and labor programs (% of total welfare of beneficiary households)",
    "Adjusted net national income per capita (current US$)",
    "Adjusted savings: education expenditure (current US$)",
    "Adolescent fertility rate (births per 1,000 women ages 15-19)",
    "Adolescents out of school (% of lower secondary school age)"
]

indicator_cols = chosen_indicators  # these are all scaled & inverted if negative

df_pca_ready = df_index.dropna(subset=indicator_cols).copy()
X = df_pca_ready[indicator_cols].values

pca = PCA()
pca.fit(X)

variance_ratios = pca.explained_variance_ratio_

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(variance_ratios)+1), variance_ratios, 'o-')
plt.title("PCA Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Variance Explained")
plt.xticks(range(1, len(variance_ratios)+1))
plt.grid(True)
plt.show()

print("PCA Explained Variance Ratios:")
for i, var_ratio in enumerate(variance_ratios, start=1):
    print(f"PC{i}: {var_ratio:.2f}")
```

```python
import matplotlib.patches as mpatches


pca_coords = pca.transform(X)

pc1_vals = pca_coords[:, 0]
pc2_vals = pca_coords[:, 1]

plt.figure(figsize=(10, 8))

sns.scatterplot(x=pc1_vals, y=pc2_vals, s=50, alpha=0.7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Biplot (Observations & Loadings)")

loadings = pca.components_.T

feature_scaling = 3
for i, indicator in enumerate(indicator_cols):

    x_load = loadings[i, 0] * feature_scaling
    y_load = loadings[i, 1] * feature_scaling

    plt.arrow(0, 0, x_load, y_load,
              color='r', alpha=0.7,
              head_width=0.02, width=0.001)
    plt.text(x_load * 1.05, y_load * 1.05, indicator, color='r', ha='center', va='center')

plt.axhline(0, color='grey', linewidth=0.8)
plt.axvline(0, color='grey', linewidth=0.8)
plt.grid(True)
plt.show()
```

```python
year_to_analyze = "2020"
df_year = df_index[df_index["Year"] == year_to_analyze].copy()

df_year.sort_values("MPI_Composite", ascending=True, inplace=True)

N = 10
bottom_n = df_year.head(N)
print(f"=== Bottom {N} Countries in {year_to_analyze} ===")
display(bottom_n[["Country Name", "MPI_Composite"] + indicator_cols])

plt.figure(figsize=(10,6))
sns.barplot(x="MPI_Composite", y="Country Name", data=bottom_n, palette="rocket")
plt.title(f"Bottom {N} Countries by MPI_Composite in {year_to_analyze}")
plt.xlabel("MPI Composite (0-1)")
plt.ylabel("Country Name")
plt.show()
```

```python
indicator_melt = bottom_n.melt(
    id_vars=["Country Name", "MPI_Composite", "Year"],
    value_vars=indicator_cols,
    var_name="Indicator",
    value_name="ScaledValue"
)

indicator_melt.sort_values(["Country Name", "ScaledValue"], ascending=[True, True], inplace=True)

print("Lowest scaled indicators for each bottom country (first 2 per country):")
low_indicators = indicator_melt.groupby("Country Name").head(2)
display(low_indicators)

plt.figure(figsize=(10, 6))
sns.barplot(x="ScaledValue", y="Country Name", hue="Indicator", data=indicator_melt, dodge=False)
plt.title(f"Indicator Comparison for Bottom {N} Countries in {year_to_analyze}")
plt.xlabel("Scaled Value (0=worst, 1=best)")
plt.ylabel("Country Name")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
def generate_policy_proposals(row):

    country = row["Country Name"]
    indicator_values = {ind: row[ind] for ind in indicator_cols}
    worst_two = sorted(indicator_values.items(), key=lambda x: x[1])[:2]

    suggestions = []
    for (ind_name, val) in worst_two:
        if "electricity" in ind_name.lower():
            suggestions.append("Improve rural electrification & power infrastructure.")
        elif "fertility" in ind_name.lower():
            suggestions.append("Expand reproductive health, education, and family planning resources.")
        elif "out of school" in ind_name.lower():
            suggestions.append("Increase school enrollment programs & reduce dropout rates.")
        elif "clean fuels" in ind_name.lower():
            suggestions.append("Promote clean cooking technologies & household energy solutions.")
        elif "social protection" in ind_name.lower():
            suggestions.append("Strengthen social safety nets & labor policies.")
        elif "net national income" in ind_name.lower():
            suggestions.append("Stimulate economic growth, job creation, or redistribution measures.")
        elif "education expenditure" in ind_name.lower():
            suggestions.append("Enhance budget allocations & efficiency in education spending.")
        else:
            suggestions.append("General multi-sector reforms.")

    policy_text = f"For {country}, focus on: " + "; ".join(suggestions)
    return policy_text

bottom_n["Policy_Proposals"] = bottom_n.apply(generate_policy_proposals, axis=1)

print("=== Tailored Policy Proposals for Bottom Countries ===")
display(bottom_n[["Country Name", "MPI_Composite", "Policy_Proposals"]])
```
